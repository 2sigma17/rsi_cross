```python
import configparser
import alpha_vantage
import requests
import pandas as pd
import numpy as np
```

```python
config = configparser.ConfigParser()
config.read('alpha_vantage.cfg')
key = config['alpha_vantage']['key']
```

```python
API_URL = 'https://www.alphavantage.co/query'
```

```python
stocks = ['TSLA']
```

```python
for stock in stocks:
    data = {
        # timeseries function
        "function": "TIME_SERIES_DAILY",
        "symbol": stock,
        # outputsize - compact = latest 100, full = 20 yrs
        "outputsize": "compact",
        "datatype": "json",
        "apikey": key
    }
    response = requests.get(API_URL, data)
    data = response.json()
    daily_ts = data['Time Series (Daily)']
#     print(f"{stock}: {daily_ts}")
```

```python
df = pd.DataFrame.from_dict(daily_ts, orient='index')
```

```python
time_periods = [50]
```

```python
for period in time_periods:
    data = {
        "function": "RSI",
        "symbol": stock,
        "interval": "daily",
        "time_period": period,
        "series_type": "close",
        "datatype": "json",
        "apikey": key
        }
    response = requests.get(API_URL, data)
    data = response.json()
    rsi_data = data['Technical Analysis: RSI']
    print(rsi_data)
```

```python
rsi = pd.DataFrame.from_dict(rsi_data, orient='index')
rsi
```

```python
df['rsi_50'] = rsi
df
```

```python
df['4. close'] = df['4. close'].astype('float')
```

```python
change = df['4. close'].diff(1)
```

```python
gain = change.mask(change<0, 0)
```

```python
df['gain'] = gain
```

```python
loss = change.mask(change>0, 0)
```

```python
df['loss'] = loss
```

```python
df
```
